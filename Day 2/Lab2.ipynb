```python
# Creating a module named `mymodule.py`
%%writefile mymodule.py

def add(a, b):
    """Function to add two numbers"""
    return a + b

def subtract(a, b):
    """Function to subtract two numbers"""
    return a - b

def multiply(a, b):
    """Function to multiply two numbers"""
    return a * b

def divide(a, b):
    """Function to divide two numbers"""
    if b == 0:
        return "Cannot divide by zero"
    return a / b
```

### Reshaping and Transposing
reshape(-1) is a line vector, when reshape(-1,1) is a column:

```python
'''
>>> a
np.array([[ 1.,  2.,  3.],
       [ 4.,  5.,  6.]])
>>> a.shape
(2, 3)
>>> a.reshape(-1)
array([ 1.,  2.,  3.,  4.,  5.,  6.])
>>> a.reshape(-1,1)
array([[ 1.],
       [ 2.],
       [ 3.],
       [ 4.],
       [ 5.],
       [ 6.]])
'''
a = np.array([[ 1.,  2.,  3.],
       [ 4.,  5.,  6.]])
print(a.shape)
print(a.reshape(-1))
print(a.reshape(-1,1))
print(a.reshape(1,-1))
```

### Pandas

### Pandas Series
*	A series is a one-dimensional structure like an associative array.
*	Series have indexes starting from 0.
*	An index refers to the position of a value in the data structure.


```python
import pandas as pd
```

### Creating a Series

```python
# From a list
data = [1, 2, 3, 4, 5]
series_from_list = pd.Series(data)
print(series_from_list)

# From a dictionary
data = {'a': 1, 'b': 2, 'c': 3}
series_from_dict = pd.Series(data)
print(series_from_dict)

# From a scalar value
series_from_scalar = pd.Series(5, index=['a', 'b', 'c'])
print(series_from_scalar)
```

### Accessing Data in a Series
* You can access elements in a Series similarly to accessing elements in a list or dictionary.

```python
# Accessing by position
print(series_from_list[2])

# Accessing by label
print(series_from_dict['a'])
```

### Vectorized Operations
*	Pandas Series supports vectorized operations, making it easy to perform element-wise operations.




```python
# Adding a scalar to each element
print(series_from_list + 10)

# Element-wise multiplication
print(series_from_list * 2)
```

### Basic Statistical Operations

```python
print(series_from_list.mean())  # Mean
print(series_from_list.sum())   # Sum
print(series_from_list.max())   # Max
print(series_from_list.min())   # Min
print(series_from_list.std())   # Standard Deviation
```

### Handling Missing Data
*	Pandas Series handles missing data elegantly using NaN (Not a Number).



```python
data = [1, 2, None, 4, 5]
series_with_nan = pd.Series(data)
print(series_with_nan)

# Checking for missing values
print(series_with_nan.isna())

# Filling missing values
print(series_with_nan.fillna(0))

# Dropping missing values
print(series_with_nan.dropna())
```

*  Calculate “Attendance Rate” = “Lecture Attendance” /13 * (2/3) + “Lab Attendance” / 13 * (1/3)
*  Define ”Grade” when


      1.   A = Mark >= 75
      2.   B = 75 > Mark >= 60  
      3.   C = 60 > Mark >= 45  
      4.    D = 45 > Mark >= 40
      5.    F = 40 > Mark or Attendance Rate < 0.8



*








```python
import pandas as pd
from google.colab import files
uploaded = files.upload() # load Data.xlsx from local drive
filename = list(uploaded.keys())[0]
data = pd.read_excel(open(filename, 'rb'), index_col = 0) # index_col to tell pandas which column (i.e. Student Name) to use as the index column when importing:
print(data)
print(data.index)
```

```python
# Print Column
print(data['Lab'])

# Print Row
print(data.loc['Mike'])
print(data.iloc[3])
print(data.iloc[:2])

data["Mark"] = data["Lab"] * 0.4 + data["Assignment"] * 0.3 + data["Final Exam"] * 0.3
data["Attendance Rate"] = data["Lecture Attendance"] / 13 * (2/3) + data["Lab Attendance"] / 13 * (1/3)

data
```

```python
print(data.dtypes)
data.loc[data["Mark"] >= 75, "Grade"] = 'A'
data.loc[(data["Mark"] >= 60) & (data["Mark"] < 75), "Grade"] = 'B'
data.loc[(data["Mark"] >= 45) & (data["Mark"] < 60), "Grade"] = 'C'
data.loc[(data["Mark"] >= 40) & (data["Mark"] < 45), "Grade"] = 'D'
data.loc[data["Mark"] < 40, "Grade"] = 'F'
data.loc[data["Attendance Rate"] < 0.8, "Grade"] = 'F'
print(data)
```
