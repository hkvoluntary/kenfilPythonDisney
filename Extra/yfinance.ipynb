```python
!pip install yahoofinancials

#import yfinance as yf # https://pypi.org/project/yfinance/
from yahoofinancials import YahooFinancials

import pandas as pd
```

```python
ticker = 'AAPL'
yahoo_financials = YahooFinancials(ticker)

balance_sheet_data_qt = yahoo_financials.get_financial_stmts('quarterly', 'balance')
balance_sheet_data_qt
```

```python
get_stock_price_data = yahoo_financials.get_stock_price_data()
get_stock_price_data
```

```python
daily_stock_prices = yahoo_financials.get_historical_price_data('2023-06-01', '2023-10-31', 'daily')
daily_stock_prices
```

```python
daily_stock_prices['AAPL']['prices']
```

```python
json_list = daily_stock_prices['AAPL']['prices']
df = pd.DataFrame.from_records(json_list)
df
```

```python
print(df.dtypes)
```

```python

df['volume']  =  df['volume'].astype('float64')
df.dtypes
```

```python
import plotly.graph_objects as go # Python classes that represent parts of a figure. - https://plotly.com/python/graph-objects/


fig = go.Figure(data=
                [go.Candlestick
                 (x=df['formatted_date'], open=df['open'], high=df['high'],
                  low=df['low'], close=df['close'])])
fig.show()
```

```python
sub_apple_df = df.iloc[-20:]
sub_apple_df
```

```python
import plotly.express as px #Plotly Express is a terse, consistent, high-level API for creating figures. - https://plotly.com/python/plotly-express/

fig = px.line(df, x="formatted_date", y="open", title='Apple Stock Prices')
fig.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize = (15,8))
ax = sns.lineplot(x='formatted_date', y='open', data=df)
ax.set_xlabel("Date")
ax.set_ylabel("Open Price")
ax.set_title("Apple Stock Prices")
plt.show()
```

```python
!pip install mplfinance

#matplotlib/mplfinance, contains a new matplotlib finance API that makes it easier to create financial plots.
#It interfaces nicely with Pandas DataFrames.
import mplfinance as fplt #
```

```python
df
```

```python
apple_df_new_index = df
apple_df_new_index.index = pd.DatetimeIndex(df['formatted_date'])
apple_df_new_index
```

```python
mc = fplt.make_marketcolors(
                            up='tab:blue',down='tab:red',
                            edge='lime',
                            wick={'up':'blue','down':'red'},
                            volume='lawngreen',
                           )

s  = fplt.make_mpf_style(base_mpl_style="seaborn", marketcolors=mc, mavcolors=["green","orange","skyblue"])

fplt.plot(
        apple_df_new_index,
        type="candle",
        title='Apple, Apple, June - October 2023',
        ylabel='Price ($)',
        mav=(2,10,20),
        figscale=2,
        style=s,
    )
```
