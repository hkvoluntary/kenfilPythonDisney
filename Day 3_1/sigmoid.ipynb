The sigmoid function is defined as:
S(x)=1/(1+e^âˆ’x)

where:

    * e is the base of the natural logarithm (approximately equal to 2.71828).
    * x is the input value.


```python
import numpy as np
import matplotlib.pyplot as plt

# Define the sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Generate input values
x = np.linspace(-10, 10, 100)
y = sigmoid(x)

# Plot the sigmoid function
plt.plot(x, y)
plt.title('Sigmoid Function')
plt.xlabel('x')
plt.ylabel('S(x)')
plt.grid()
plt.axhline(0, color='black', lw=0.5)
plt.axvline(0, color='black', lw=0.5)
plt.show()
```

sigmoid function itself always outputs values strictly between 0 and 1, meaning its range is (0, 1).


Scaling the Sigmoid Output

You can scale the output of the sigmoid function to any desired range [a,b]using the following formula:
y=a+(bâˆ’a)â‹…S(x)

Where S(x) is the sigmoid output.



```python
import numpy as np
import matplotlib.pyplot as plt

def scaled_sigmoid(x, a=-1, b=1):
    return a + (b - a) * (1 / (1 + np.exp(-x)))

# Generate input values
x = np.linspace(-10, 10, 100)
y = scaled_sigmoid(x)

# Plot
plt.plot(x, y)
plt.title('Scaled Sigmoid Function to Range [-1, 1]')
plt.xlabel('x')
plt.ylabel('Scaled S(x)')
plt.grid()
plt.axhline(0, color='black', lw=0.5)
plt.axvline(0, color='black', lw=0.5)
plt.show()
```
